```python
def featurePlotCreator(i):
    feature1 = []
    feature2 = []
    feature3 = []
    feature4 = []
    feature5 = []
    feature6 = []

    idx = np.where(y_train == i)
    samples = x_train[idx, :, :]
    samples = np.concatenate([x for x in samples[0] ], axis=0)

    for x in samples:
        feature1.append(x[0])
        feature2.append(x[1])
        feature3.append(x[2])
        feature4.append(x[3])
        feature5.append(x[4])
        feature6.append(x[5])

    figure, axis = plt.subplots(2, 3, figsize=(20,10))
    
    figure.suptitle(f'CLASS {i+1}')

    axis[0, 0].plot(feature1)
    axis[0, 0].set_title("Feature one")

    axis[0, 1].plot(feature2)
    axis[0, 1].set_title("Feature two")

    axis[0, 2].plot(feature3)
    axis[0, 2].set_title("Feature three")

    axis[1, 0].plot(feature4)
    axis[1, 0].set_title("Feature four")

    axis[1, 1].plot(feature5)
    axis[1, 1].set_title("Feature five")

    axis[1, 2].plot(feature6)
    axis[1, 2].set_title("Feature six")

    plt.show()
    
```

```python
for i in range(12):
    featurePlotCreator(i)
```
